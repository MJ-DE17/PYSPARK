```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col , length
from pyspark.sql.functions import *

spark = SparkSession.builder.appName("String funcitno").getOrCreate()

data = [
    (1, "9Manasa devi", 50000.0, 25),
    (2, "Mevin", 60000.0, 28),
    (3, "2Pavithra ", 55000.0, 26),
    (4, "1Menakha", 70000.0, 30),
    (5, "Arjun ", 45000.0, 24)
]

columns = ["id" , "name" , "salary " , "age"]

df = spark.createDataFrame(data , columns)
```

```python
df.select(col("name") , upper(col("name")).alias("uppercase") , lower("name").alias("Lowercase_names") , initcap("name").alias("title")).show()
```

```python
# Trimming & Cleaning
#  trim canbe used only with the select & withcolumn

df.select(col("name") , trim(col("name")).alias("Nor trim") , ltrim(col("name")).alias("L trim"),rtrim(col("name")).alias("R trim")).show()
```

```python
# Length & Position

# length
# instr - return count of the substring found.

df.select(trim("name").alias("name"))\
.filter(length(col("name")) <= 5).show()
df.select("name" , instr("name" , "na").alias("SLiced")).show()
# df.filter(instr(col("name" ), "na") > 0).show()
```

```python
# Substring & Extraction

# substring- prints part of strng
# split - print separated list (word , (element wehre to split))

df.select("name" , substring("name" , 1 , 6), split("name" , "i") ).show()
df.select(
    "name",
    substring("name", 1, 6).alias("sub_name"),
    split("name", "i").alias("split_name")
).show(truncate=False)

# regexp_extract

df.select("name" , regexp_extract("name" , r"^M", 0)).show()
```

```python
# Replacement & Cleaning

# Function	Purpose
# replace()	Replace substring
# regexp_replace()	Replace using regex
# translate()	Replace characters one-to-one

df.select("name" , regexp_replace("name" , "[\\d]" , "").alias("no_number")).show()
df.select("name" , regexp_replace(col("name") , "a" , "A").alias("no_number")).show()
```

```python
# Concatenation

# concat()	- Join columns
# concat_ws()	- Join with separator

df.select(
    trim(col("name")).alias("name"),
    concat(col("name"), col("age")).alias("password"),
    concat_ws("_", col("name"), col("age")).alias("username")
).show()
```

```python
# Pattern Matching

# like()	SQL pattern -Case-sensitive
# ilike()	Case-insensitive
# rlike()	Regex - Case-insensitive


df.filter(col("name").like("M%")).show()
df.filter(col("name").like("%a")).show()
df.filter(col("name").rlike("^M")).show() #startwith
df.filter(col("name").rlike("n$")).show() #endswith
df.filter(col("name").rlike("^[A-Za-z]+$")).show() #
df.filter(col("name").rlike("[0-9]")).show()  # or [^\\d] Names starting with number


```

```python
# formatting, masking, and string shaping.

df.select(
    "name",
    translate("name", "a", "A").alias("translated")
).show()


# Remove digits
df.select(
    "name",
    regexp_replace("name", r"[0-9]", "").alias("no_digits")
).show()
print("-----M")
# This replaces whole value "Manasa" â†’ "M"
df.replace("Mevin", "M").show()

# combined other methods too
df.select(
    "name",
    repeat(
         regexp_replace(trim("name") , r"^\d+", ""),
         2
    ).alias("Repeatation")
).show()
```

```python
# rpad
# Pads string on the right side until it reaches a fixed length.
# rpad(column, length, pad_string)

# !!If string length > specified length ,
# Also truncates if original string is longer.

# rpad : Raj_______
df.select("name" , rpad("name" , 10 , "_").alias("rpad_name")).show()

# lpad : _____Mevin
df.select("name" , lpad("name" , 16 , "*").alias("lpad_name")).show()

```
