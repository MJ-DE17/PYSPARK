```python
# udf :


from pyspark.sql import SparkSession
from pyspark.sql.functions import udf, col
from pyspark.sql.types import *

spark = SparkSession.builder.appName("UDF_Practice").getOrCreate()

data = [
    (1, "Manasa", 22),
    (2, "Mevin", 35),
    (3, "Pavithra", 41),
    (4, "Menakha", 28)
]

columns = ["id", "name", "age"]

df = spark.createDataFrame(data, columns)
df.show()

@udf(StringType())
def age_udf(age):
  if age < 25 :
    return "Young"
  elif age < 35:
    return "mid age"
  else :
    return "old age"

df_new = df.withColumn("Age Category " , age_udf(col("age")))
df_new.show()
```

```python
# Bonus Calculation

data = [
    (1, "Manasa", 60000),
    (2, "Mevin", 45000),
    (3, "Pavithra", 30000)
]

columns = ["id", "name", "salary"]

df = spark.createDataFrame(data, columns)
df.show()

@udf(DoubleType())
def bonus_udf(salary):
  if salary <= 35000 :
    return salary + (salary * 0.1)
  elif salary <= 45000 :
    return salary + (salary * 0.2)
  else :
    return salary + (salary * 0.3)

new_df = df.withColumn("Increased salary" , bonus_udf(col("salary")))

new_df.show()
```
