```python
# csv

# Method 1: Using inferSchema (Beginner way)

from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType # Added this import

spark = SparkSession.builder \
    .appName("ReadCSV") \
    .getOrCreate()

df = spark.read.csv("/content/industry.csv" , header = True , inferSchema = True)
df.printSchema()
df.show()


# Method 2: Define Schema Manually (Professional Way)


schema = StructType([StructField("Industryy" , StringType(), True)])

sf_1 = spark.read \
       .option("header" ,"true") \
       .schema(schema) \
       .csv("/content/industry.csv")

sf_1.printSchema()
sf_1.show()
```

```python
# parque

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ReadParquet") \
    .getOrCreate()

data = [(1, "Manasa", 23),
        (2, "Mevin", 25)]

columns = ["id", "name", "age"]

df = spark.createDataFrame(data, columns)

# write first , so te parque folder si screated and then we can read it .

df.write.mode("overwrite").parquet("Parquet_file")

df.show()

df2 = spark.read.parquet("Parquet_file")
df2.printSchema()
df2.show()
```

```python
# tables , internal

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Tables").enableHiveSupport().getOrCreate()


data = [
    (1, "Ravi", "Chennai"),
    (2, "Anu", "Salem")
]

columns = ["id", "name", "city"]

df = spark.createDataFrame(data, columns)

df.write.mode("overwrite").saveAsTable("emp_managed")
```
